```python
import pandas as pd

table = pd.read_excel("/Users/nguyenquangminh/Downloads/Điểm-học-phần.xlsx", header=None)
table.columns = ["Semester", "Course ID","Tên tiếng việt", "Tín chỉ", "Điểm chữ", "Tên tiếng anh"]
def filter_mon_phu(row):
    if row["Tên tiếng việt"] in ["QS chung và KCT bắn súng AK", "Đường lối quân sự", "Công tác quốc phòng-An ninh"]:
        return True
    return False
def change_score(row):
    if row["Điểm chữ"] in ["A", "A+"]:
        return 4
    elif row["Điểm chữ"] == "B+":
        return 3.5
    elif row["Điểm chữ"] == "B":
        return 3
    elif row["Điểm chữ"] == "C+":
        return 2.5
    elif row["Điểm chữ"] == "C":
        return 2
    elif row["Điểm chữ"] == "D+":
        return 1.5
    elif row["Điểm chữ"] == "D":
        return 1
    elif row["Điểm chữ"] == "F":
        return 0
table =  table[table["Tên tiếng việt"] != "QS chung và KCT bắn súng AK"]
table =  table[table["Tên tiếng việt"] != "Đường lối quân sự"]
table =  table[table["Tên tiếng việt"] != "Công tác quốc phòng-An ninh"]
table =  table[table["Tên tiếng việt"] != "Chạy"]
table =  table[table["Tên tiếng việt"] != "Erobic"]
table =  table[table["Tên tiếng việt"] != "Bóng bàn 1"]
table =  table[table["Tên tiếng việt"] != "Lý luận TDTT"]
table =  table[table["Tên tiếng việt"] != "Cầu lông 1"]
table =  table[table["Điểm chữ"] != "R"]
table["Score"] = table.apply(change_score, axis=1)
table['product'] = table.apply(lambda row: row['Tín chỉ'] * row['Score'], axis=1)
print(table["product"].sum() / table["Tín chỉ"].sum())
print(table)
```
